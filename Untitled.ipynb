```python
import pandas as pd
import os
```

```python
new_df = pd.DataFrame()

for file in os.listdir('google_data/company_info/'):
    test_df = pd.read_csv(f'google_data/company_info/{file}')
    new_df = pd.concat([new_df, test_df])
```

```python
new_df.drop_duplicates(inplace=True)
```

```python
new_df.to_csv('BASE_INFO.csv', index=False)
```
